```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('911.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['zip'].value_counts().head(5)
```

```python
df['twp'].value_counts().head(5)
```

```python
df['title'].nunique()
```

```python
df['Reason'] = df['title'].apply(lambda x: x.split(':')[0])   
df
```

```python
df['Reason'].value_counts().head(1)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Reason', data=df, palette='rainbow')
```

```python
type(df['timeStamp'].iloc[3])
```

```python
timestamp = pd.to_datetime(df['timeStamp'])
timestamp
```

```python
df['Hour'] = timestamp.apply(lambda x: x.hour)
df['Month'] = timestamp.apply(lambda x: x.month)
df['Day of Week'] = timestamp.apply(lambda x: x.dayofweek)
df
```

```python
time = timestamp.iloc[3]
time
```

```python
time.dayofweek
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
days = df['Day of Week'].map(dmap)
days
```

```python
df['Day of Week'] = days
df
```

```python
sns.countplot(x='Day of Week', data = df, hue = 'Reason', palette = 'rainbow').legend(bbox_to_anchor=(1.05,1), loc='upper left')
```

```python
sns.countplot(x='Month',data=df,hue='Reason',palette='rainbow').legend(bbox_to_anchor=(1.05,1), loc='upper left')
```

```python
byMonth = df.groupby('Month').count()
byMonth.head()
```

```python
byMonth['lat'].plot()
```

```python
byMonth.reset_index(inplace= True)
```

```python
sns.lmplot(data= byMonth, x= 'Month', y= 'twp')
```

```python
dmap = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
months = df['Month'].map(dmap)
df['Months']= months
df
```

```python
del df['Date']
df
```

```python
df['Date'] = timestamp.apply(lambda x: x.date())
df['Date']
```

```python
byDate = df.groupby('Date').count()
byDate['lat'].plot()
```

```python
df[df['Reason']== 'Traffic'].groupby('Date').count()['lat'].plot()
plt.title('Traffic')
```

```python
df[df['Reason']== 'Fire'].groupby('Date').count()['lat'].plot()
plt.title('Fire')
```

```python
df[df['Reason']== 'EMS'].groupby('Date').count()['lat'].plot()
plt.title('EMS')
```

```python
s = df.groupby(by=['Day of Week','Hour']).count()['lat'].unstack()
s.head()
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(data=s, cmap='coolwarm')
```

```python
sns.clustermap(data=s, cmap='coolwarm')
```

```python
m = df.groupby(['Day of Week','Months']).count()['lat'].unstack()
m.head()
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(data=m, cmap='coolwarm')
```

```python
sns.clustermap(data=m, cmap='coolwarm')
```
